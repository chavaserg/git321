Вивід трапеції на пайтоні

```python
import matplotlib.pyplot as plt

# Визначення координат вершин трапеції
x = [1, 3, 5, 2]  # x-координати
y = [1, 1, 4, 3]  # y-координати

# Створення графіка
plt.plot(x + [x[0]], y + [y[0]], marker='o')

# Налаштування відображення
plt.xlabel('X-координата')
plt.ylabel('Y-координата')
plt.title('Трапеція')

# Відображення графіка
plt.show()
```
